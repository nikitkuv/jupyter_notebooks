# Восстановление золота из руды

Предоставлены данные с параметрами добычи и очистки золотосодержащей руды от компании, занимающейся разработкой решения для эффективной работы промышленных предприятий.

**Цель:** подготовить прототип модели машинного обучения, предсказывающую коэффициент восстановления золота из золотосодержащей руды.

**Целевой признак:** коэффициент восстановления золота в черновом и финальном концентратах (rougher.output.recovery + final.output.recovery).

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

# Описание данных

Технологический процесс:
   - Rougher feed — исходное сырье
   - Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
   - Xanthate — ксантогенат (промотер, или активатор флотации);
   - Sulphate — сульфат (на данном производстве сульфид натрия);
   - Depressant — депрессант (силикат натрия).
   - Rougher process (англ. «грубый процесс») — флотация
   - Rougher tails — отвальные хвосты
   - Float banks — флотационная установка
   - Cleaner process — очистка
   - Rougher Au — черновой концентрат золота
   - Final Au — финальный концентрат золота

Параметры этапов:
   - air amount — объём воздуха
   - fluid levels — уровень жидкости
   - feed size — размер гранул сырья
   - feed rate — скорость подачи

Наименование признаков:
 - Наименование признаков должно быть такое: *этап.тип_параметра.название_параметра*
 - Пример: rougher.input.feed_ag

Возможные значения для блока *этап*:
   - rougher — флотация
   - primary_cleaner — первичная очистка
   - secondary_cleaner — вторичная очистка
   - final — финальные характеристики

Возможные значения для блока *тип_параметра*:
   - input — параметры сырья
   - output — параметры продукта
   - state — параметры, характеризующие текущее состояние этапа
   - calculation — расчётные характеристики

## Введение

### Расчет эффективности

Эффективность обогащения рассчитывается по формуле:

![title](recovery.jpg)

где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

### Метрика качества

Для решения задачи будем использовать метрику качества sMAPE.

Метрика sMAPE вычисляется так:

![title](smape.jpg)

Нужно спрогнозировать сразу две величины:
- эффективность обогащения чернового концентрата rougher.output.recovery;
- эффективность обогащения финального концентрата final.output.recovery.

Итоговая метрика складывается из двух величин:

### Проверка признака recovery

- Проверим исходные значения признака recovery: сравним их с посчитанными по формуле значениями
- Для этого сделаем функцию для подсчета эффективности обошащения

```python
gold_train['final.output.recovery'].describe()
```

```python
def recovery(row):
    C = row['final.output.concentrate_au']
    F = row['rougher.input.feed_au']
    T = row['final.output.tail_au']
    return (C*(F-T))/(F*(C-T))*100
```

- Посчитаем значения по написанной функции и сравним основные статистические параметры

```python
gold_train['recovery_custom'] = gold_train.apply(recovery, axis=1)
gold_train['recovery_custom'].describe()
```

```python
gold_train['final.output.recovery'].describe()
```

- Дополнительно сравним оба распределения

```python
fig, ax = plt.subplots(figsize=(8, 10))

sns.boxplot(x='calculation',
            y='recovery',
            data=pd.DataFrame({'recovery':pd.concat([gold_train['final.output.recovery'], 
                                                     gold_train['recovery_custom']
                                                    ],
                                                    axis=0),
                               'calculation':np.repeat(['Original',
                                                        'Custom'
                                                        ],
                                                         [len(gold_train['final.output.recovery']),
                                                          len(gold_train['recovery_custom'])
                                                         ]
                                                        ),
                              }
                             ),
            #whis=[2.5, 97.5]
           )

plt.title('Recovery calculation check')
plt.show()
```

- Посчитаем среднюю абсолютную ошибку

```python
recovery_mae = sum(abs(gold_train['final.output.recovery'] - gold_train['recovery_custom']))/len(gold_train['final.output.recovery'])
recovery_mae
```

### Создание функций для метрик

- Для начала напишем функцию для расчета метрики sMAPE

```python
def smape(target, predictions):
    """
    calculates final sMAPE metric
    
    params:
    - target - 2-dimension target array
    - predictions - 2-dimension predictions array
    
    returns:
    - r - rougher sMAPE
    - f - final sMAPE
    """
    r = 1/len(target[:,0])*np.sum(2*np.abs(target[:,0]-predictions[:,0])/(np.abs(target[:,0])+np.abs(predictions[:,0]))*100)
    f = 1/len(target[:,1])*np.sum(2*np.abs(target[:,1]-predictions[:,1])/(np.abs(target[:,1])+np.abs(predictions[:,1]))*100)
    return r, f
```

```python
def final_smape(target, predictions):
    """
    calculates final sMAPE metric
    
    params:
    - rsmape - sMAPE on rougher output
    - fsmape - sMAPE on final output
    
    returns:
    - final sMAPE value
    """
    return 0.25*smape(target, predictions)[0]+0.75*smape(target, predictions)[1]
```

- Для подбора гиперпараметров с использованием кросс-валидации будем сразу использовать sMAPE как метод скоринга

```python
final_smape_scorer = make_scorer(final_smape, greater_is_better=False)
```

- Теперь напишем функцию, которая, принимая на выход обучающую и тестовую выборки, обучала модель, делала предсказания и оценивала качество модели с помощью кросс-валидации

- Создадим функцию для проверки разных моделей

```python
def reg_gs_score(model, x_train, y_train, x_test, y_test, cv_k, hp_space, seed=24, scale=False, randomcv=False):
    """
    scores the regression model using RandomizedSearchCV, final sMAPE metric and R squared
    
    params:
    - model - the regression model
    - x_train - training features
    - y_train - training labels
    - x_test - test features
    - y_test - test labels
    - cv_k - k for k-fold cross-validation in RandomizedSearchCV
    - hp_space - hyperparameter space
    - seed - seed
    - scale=False - True if the model needs to be scaled (StandardScaler applied)
    
    returns:
    - final_smape_test - final sMAPE on the test set
    - reg_gs_validation_score - final sMAPE on the validation set
    - r2_test - r squared on the test set
    - reg_gs_best_params - model's best parameters from RandomizedSearchCV based on final sMAPE
    - reg_gs_best_model - best model from RandomizedSearchCV based on final sMAPE
    """
    if scale:
        scaler = StandardScaler()
        reg = model
        
        steps = [('scaler', scaler), ('reg', reg)]
        pipeline = Pipeline(steps)
        
        new_hp_space = {}
        for key, val in hp_space.items():
            new_hp_space['reg__'+key] = val
            
        hp_space = new_hp_space.copy()
    else:
        pipeline = model
        
    if randomcv:
        reg_gs = RandomizedSearchCV(pipeline, 
                                    hp_space, 
                                    cv=cv_k, 
                                    scoring=final_smape_scorer,
                                    random_state=seed
                                   )
    else:
        reg_gs = GridSearchCV(pipeline, 
                              hp_space, 
                              cv=cv_k, 
                              scoring=final_smape_scorer,
                             )
    
    reg_gs.fit(x_train, y_train)
    
    reg_gs_best_params = reg_gs.best_params_
    reg_gs_best_model = reg_gs.best_estimator_
    reg_gs_validation_score = -reg_gs.best_score_.round(3)
    
    if scale:
        reg_gs_best_params_out_of_pipe = {}
        for key, val in reg_gs_best_params.items():
            reg_gs_best_params_out_of_pipe[key[5:]] = val
    else:
        reg_gs_best_params_out_of_pipe = reg_gs_best_params.copy()
    
    y_pred = reg_gs_best_model.predict(x_test)
    final_smape_test = final_smape(y_test, y_pred).round(3)
    r2_test = r2_score(y_test, y_pred).round(3)
    
    return final_smape_test, reg_gs_validation_score, r2_test, reg_gs_best_params, reg_gs_best_params_out_of_pipe, reg_gs_best_model
```

### Базовые модели: LinearRegression

- Для начала рассмотрим линейную регрессию в качестве базовой модели
- Попробуем два варианта:
    1. Модель со всеми исходными признаками
    2. Модель без признаков, для которых коэффициенты в Lasso равны 0

- Напишем функцию для оценки базовых моделей линейной регрессии с использованием кросс-валидации и метрик final sMAPE и R squared

```python
r2_scorer = make_scorer(r2_score, greater_is_better=True)
```

```python
def bm_linreg_cv_score(x_train, y_train, x_test, y_test, cv_k):
    """
    
    """
    scaler = StandardScaler()
    linreg = LinearRegression()

    steps = [('scaler', scaler), ('linreg', linreg)]
    pipeline = Pipeline(steps)

    train_val_score_linreg_1 = cross_validate(pipeline,
                                              x_train, 
                                              y_train, 
                                              cv=cv_k, 
                                              scoring={'fsmape': final_smape_scorer,
                                                       'r2': r2_scorer  
                                                      },
                                              return_train_score=True
                                              )

    pipeline.fit(x_train, y_train)
    y_pred = pipeline.predict(x_test)

    fsmape_test = round(final_smape(y_test, y_pred), 3)
    r2_test = round(r2_score(y_test, y_pred), 3)
    
    fsmape_val = -round(1/((train_val_score_linreg_1['test_fsmape']).mean()), 3)
    r2_val = round(train_val_score_linreg_1['test_r2'].mean(), 3)
    
    fsmape_train = -round(1/((train_val_score_linreg_1['train_fsmape']).mean()), 3)
    r2_train = round(train_val_score_linreg_1['train_r2'].mean(), 3)
    
    return fsmape_test, r2_test, fsmape_val, r2_val, fsmape_train, r2_train
```

**Базовая модель 1: все признаки**

```python
fsmape_test_bm1, r2_test_bm1, fsmape_val_bm1, r2_val_bm1, fsmape_train_bm1, r2_train_bm1 = bm_linreg_cv_score(x_train, 
                                                                                                              y_train, 
                                                                                                              x_test, 
                                                                                                              y_test, 
                                                                                                              5
                                                                                                              )

print(f'final sMAPE test: {fsmape_test_bm1}')
print(f'R2 test: {r2_test_bm1}')
print('')
print(f'final sMAPE valid: {fsmape_val_bm1}')
print(f'R2 valid: {r2_val_bm1}')
print('')
print(f'final sMAPE train: {fsmape_train_bm1}')
print(f'R2 train: {r2_train_bm1}')
```
