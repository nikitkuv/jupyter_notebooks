# Анализ закономерностей успешности видеоигр

Рассматриваются исторические данные о продажах видеоигр, оценки пользователей и экспертов, жанры и платформы.

**Цель:**  Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

# Описание данных

 - Name — название игры
 - Platform — платформа
 - Year_of_Release — год выпуска
 - Genre — жанр игры
 - NA_sales — продажи в Северной Америке (миллионы проданных копий)

- Для части строк с пропусками в столбце year_of_release год выпуска по сути написан в названии игры;
- Запись года в названии игры как правило характерна для игр спортивного жанра (в данных также в жанре Racing);
- Можно заметить, что помимо традиционной записи года в названии с использованием четырехзначных чисел (FIFA Soccer 2004), также есть игры, где в названии указаны только две последние цифры года (Triple Play 99);
- Создадим функции для извлечения года из названия игры для этих двух вариантов;

```python
def parsing_year_from_name_four_digits(value):
    """
    извлекает четырехзначное обозначение года из названия игры
    """
    try:
        return int(value[-4:]) if int(value[-4:]) < 2017 and int(value[-4:]) > 1979 else 0
    except:
        return 0
```

```python
def parsing_year_from_name_two_digits(row):
    """
    на основе извлеченного двухначного обозначения года возвращает четырехзначное обозначение года
    """
    try:
        if row['year_from_name'] == 0 and int(row['name'][-3:]) < 100 and int(row['name'][-3:]) > 79:
            return int('19'+row['name'][-2:])
        elif row['year_from_name'] == 0 and int(row['name'][-3:]) < 17:
            return int('20'+row['name'][-2:])
        else:
            return row['year_from_name']
    except:
        return 0
```

```python
games['year_from_name'] = games['name'].apply(parsing_year_from_name_four_digits)
games['year_from_name'].value_counts()
```

```python
games['year_from_name_2_digits'] = games.apply(parsing_year_from_name_two_digits, axis=1)
games.loc[games['year_of_release'].isna(), 
         ['name', 'year_of_release', 'year_from_name', 'year_from_name_2_digits']
         ].head(15)
```

- Теперь создадим функцию для объеднения результатов двух столбцов с извлеченными обозначениями года;

```python
def year_columns_merge(row):
    """
    объединяет значения столбцов с результатами извлечения обозначения года из названия игр
    """
    if row['year_from_name_2_digits'] > row['year_from_name']:
        return row['year_from_name_2_digits']
    else:
        return row['year_from_name']
```

```python
games['year_from_name'] = games.apply(year_columns_merge, axis=1)
games.drop(columns=['year_from_name_2_digits'], inplace=True)
games[games['year_of_release'].isna()].head(15)
```

- Разберемся сначала со значениями "tbd" (-1) в столбце user_score. TBD - To Be Determined - ставится в том случае, если количество оценок не достигло определенного требуемого минимального. Сделано это для того, чтобы получить более объективную оценку и снизить вероятность влияния на оценку слишком низких или слишком высоких единичных значений;
- Выделим отдельный датасет без пропусков в столбцах critic_score и user_score и без значений "tbd" в столбце user_score. Затем посмотрим зависимость user_score от critic_score и, если она есть, заменим значения "tbd" на значения, вычисленные с помощью значений столбца critic_score по уравнению регрессии;

```python
no_tbd_no_nans_in_scores = games[(~games['critic_score'].isna()) & 
                                 (~games['user_score'].isna()) & 
                                 (games['user_score'] != -1)
                                ]
```

```python
no_tbd_no_nans_in_scores['user_score'].describe()
```

```python
slope, intercept = np.polyfit(no_tbd_no_nans_in_scores['critic_score'], 
                              no_tbd_no_nans_in_scores['user_score'], 
                              1
                             )
pearson = no_tbd_no_nans_in_scores['critic_score'].corr(no_tbd_no_nans_in_scores['user_score'])

no_tbd_no_nans_in_scores.plot(kind='scatter', 
                   x='critic_score', 
                   y='user_score', 
                   alpha=0.5, 
                   grid=True
                    )
plt.plot(np.linspace(0, 100), 
         np.linspace(0, 100)*slope + intercept, 
         #linestyle=':', 
         color='red'
        )
plt.text(80, 1, f'Pearson: {round(pearson, 2)}')
plt.show()
```

- Уравнение линейной регрессии имеет не очень высокий коэффициент корреляции, что видно визуально;
- Прослеживается нелинейная зависимость, побробуем построить полином 2-ой степени;

```python
coeffs_user = np.polyfit(no_tbd_no_nans_in_scores['critic_score'], 
                         no_tbd_no_nans_in_scores['user_score'], 
                         2
                        )

no_tbd_no_nans_in_scores.plot(kind='scatter', 
                              x='critic_score', 
                              y='user_score', 
                              alpha=0.5, 
                              grid=True
                             )
plt.plot(np.linspace(0, 100), 
         (np.linspace(0, 100)**2)*coeffs_user[0] + np.linspace(0, 100)*coeffs_user[1] + coeffs_user[2], 
         #linestyle=':', 
          color='red'
         )

plt.show()
```

- Визуально график полинома 2-ой степени ложится лучше на данные;
- Посмотрим, сколько значений мы можем таким образом заполнить: не все строки с "tbd" в столбце user_score имеют значения в столбце critic_score, часть из них имеет пропуски;
- Кроме того, часть строк с tbd в user_score имеют пропуски в столбце critic_score;

- Таким образом, чуть меньше половины значений tbd в user_score мы можем заменить на значение в соответствии с уравнением регрессии;
- Посмотрим на распределения оценок критиков для строк со значениями tbd в столбце оценок пользователей и на остальные;

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.distplot(games[(~games['critic_score'].isna()) & 
                   (games['user_score'] == -1)]['critic_score'], 
             label='Critic score for tbd',
             bins=30
            )
sns.distplot(games[(~games['critic_score'].isna()) & 
                   (games['user_score'] != -1)]['critic_score'], 
             label='Critic score for no tbd',
             bins=30
            )
plt.legend(loc=2)
plt.show()
```

- Мы видим, что значения оценок критиков для игр, где оценки пользователей равны tbd, смещены в сторону меньших оценок;

- Для начала сравним, изменится ли распределение оценок пользователей после замены значений tbd в зависимости от оценок критиков в своответствии с уравнением регрессии;

```python
# считаем количество лет, которое платформа находилась на рынке
sales_by_platform['total_years'] = games.pivot_table(index='platform', 
                                                     values='year_of_release', 
                                                     aggfunc=lambda x: x.max() - x.min()
                                                     )['year_of_release']

# считаем какое количество игр в среднем продается на платформе
sales_by_platform['total_sales_per_year'] = sales_by_platform['total_sales']/sales_by_platform['total_years']

# создаем функцию для обработки пропусков для платформ, образовавшихся в 2016 году
def nans_handling(row):
    if row['total_years'] == 0:
        return row['total_sales']
    else:
        return row['total_sales_per_year']

# обрабатываем пропуски и сортируем результаты
sales_by_platform['total_sales_per_year'] = sales_by_platform.apply(nans_handling, axis=1)
sales_per_year_by_platform_sorted = sales_by_platform.sort_values(by=['total_sales_per_year'], 
                                                                  ascending=False
                                                                 )

# выводим распределение продаж удельно за год по платформам
plt.figure(figsize=(10,6))
bar = sns.barplot(x=sales_per_year_by_platform_sorted.index, 
            y=sales_per_year_by_platform_sorted['total_sales_per_year'],
            #palette='YlGnBu_r'
            palette='gist_earth'
           )

# выведем в качетстве подписи столбика количество лет на рынке
for index, row in sales_per_year_by_platform_sorted.reset_index().iterrows():
    bar.text(row.name, 
             row.total_sales_per_year, 
             row.total_years, 
             ha='center',
            )

plt.title('Yearly total sales per platform with total years on the market')
plt.xticks(rotation=90)
plt.show()
```

```python
# создаем таблицу для извлечения года пика = года максимальных продаж
gtp = games_top_platforms.pivot_table(index='year_of_release', 
                                      values='total_sales', 
                                      columns='platform',
                                      aggfunc='sum'
                                     )
gtp.fillna(0, inplace=True)
```

```python
# извлекаем год пика
max_sale_year = []

for name, column in gtp.reset_index().iteritems():
    if name != 'year_of_release':
        max_sale_year.append(gtp.reset_index()[gtp.reset_index()[name] == column.max()]['year_of_release'].values[0])
```

```python
# добавляем в таблицу продаж по годам и по платформам столбцы года пика и считаем годы до пика
tpss = sales_per_year_by_platform_sorted[sales_per_year_by_platform_sorted['total_sales_per_year'] > 50].sort_index()
tpss['max_sale_year'] = max_sale_year
tpss['years_to_the_peak'] = tpss['max_sale_year'] - tpss['start_year']

# пишем функцию для пометки новых платформ (живут меньше 5 лет)
def five_years_interval(row):
    if (2016-row['start_year']) < 5:
        return np.nan
    else:
        return row['years_to_the_peak']

# удаляем новые платформы, меняет тип данных и сортируем по годам до пика (5 лет - время до пика продаж)
tpss['years_to_the_peak'] = tpss.apply(five_years_interval, axis=1)
tpss.dropna(inplace=True)
tpss['years_to_the_peak'] = tpss['years_to_the_peak'].astype('int')

tpss = tpss.sort_values(by=['total_years'])

# выводим распределение общему времени жизни платформ
fig, ax = plt.subplots(figsize=(10,6))
bar_1 = sns.barplot(x=tpss.index,
                    y=tpss['total_years'],
                    palette='gist_earth'
                   )
plt.title('Total life span for >= 5 years old platforms')

# выведем в качетстве подписи столбика количество лет на рынке
for index, row in tpss.reset_index().iterrows():
    bar_1.text(row.name, 
               row.total_years, 
               row.total_years, 
               ha='center',
               )

# выводим распределение по годам до пика с общим временем жизни платформы
fig, ax = plt.subplots(figsize=(10,6))
bar_2 = sns.barplot(x=tpss.index,
            y=tpss['years_to_the_peak'],
            palette='gist_earth'
           )
plt.title('Years to sales peak for >= 5 years old platforms')

# выведем в качетстве подписи столбика количество лет до пика
for index, row in tpss.reset_index().iterrows():
    bar_2.text(row.name, 
               row.years_to_the_peak, 
               row.years_to_the_peak, 
               ha='center',
               )

plt.show()
```

- На графиках видно, что время жизни топ платформы - в среднем 9-11 лет. При этом пик продаж наступает в среднем от 3 до 5 лет. Есть одно исключение - 3DS - которая на рынке уже 5 лет (и продолжает существовать, пускай и на спаде продаж игр) и при этом пик продаж был в год выпуска платформы, что скорее является случайным событием, нежели существующим трендом;
- Среднее время жизни платформы - 9-11 лет, при это в среднем на пик продаж платформ выходит 3-5 лет. Исходя из этих данных, платформы, появившиеся 5 лет назад, должны находиться сейчас на пике продаж, либо в скором времени на него выходить;

## Проверка гипотез

### Анализ пользовательских рейтингов

- У нас есть две гипотезы:
    1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    2. Средние пользовательские рейтинги жанров action и sports разные;

- Начнем с первой. Определелим нулевую и альтернативную гипотезы:
    - Нулевая гипотеза H0: Средние пользовательские рейтинги платформ Xbox One и PC **равны**;
    - Альтернативная гипотеза H1: Средние пользовательские рейтинги платформ Xbox One и PC **не равны**;
- Нам нужно проверить равенство средние двух выборок, т.е. нам нужно понять, принадлежат ли выборке одной генеральной совокупности, или двум разным;
- Подходящим методом будет парный т-тест, но для начала проверим распределение на нормальность и посмотрим количество элементов в выборках, чтобы понять, можно ли применять т-тест;

```python
# для проверки первой гипотезы уберем все пропуски и значения tbd
first_hypothesis = games_actual.copy(deep=True)
first_hypothesis = first_hypothesis[['platform', 'user_score']]

print('Dataset length BEFORE', first_hypothesis.shape[0])

first_hypothesis['user_score'].replace(-1, np.nan, inplace=True)
first_hypothesis.dropna(subset=['user_score'], inplace=True)

print('Dataset length AFTER', first_hypothesis.shape[0])
print('')
print('Number of NaNs: ')
display(first_hypothesis.isna().sum())
print('Number of tbd values:', first_hypothesis[first_hypothesis['user_score'] < 0]['user_score'].count())
```

```python
# создаем отдельные датасеты для двух платформ
xone = first_hypothesis[first_hypothesis.platform == 'XOne']['user_score']
pc = first_hypothesis[first_hypothesis.platform == 'PC']['user_score']

# делаем стандартизацию
xone_z = (xone-np.mean(xone))/np.std(xone)
pc_z = (pc-np.mean(pc))/np.std(pc)

# выводим Q-Q графики
fig, ax = plt.subplots(figsize=(10, 6))
fig_1, ax_1 = plt.subplots(figsize=(10, 6))

st.probplot(xone_z, dist='norm', plot=ax)
st.probplot(pc_z, dist='norm', plot=ax_1)

ax.set_title('Normal Q-Q plot Xbox One')
ax_1.set_title('Normal Q-Q plot PC')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.distplot(xone, 
             label='Xbox One',
             bins=15
            )
sns.distplot(pc, 
             label='PC',
             bins=15
            )

plt.ylim(0, 0.4)
plt.axvline(xone.mean(), 
            c='tab:blue', 
            linewidth=1.5, 
            linestyle='--'
           )
plt.axvline(pc.mean(), 
            c='peru', 
            linewidth=1.5, 
            linestyle='--'
           )
ax.text(xone.mean()+.1, 
        0.328, 
        'XOne mean', 
        rotation=90
        )
ax.text(pc.mean()-.2, 
        0.345, 
        'PC mean', 
        rotation=90
        )


plt.legend(loc=2)
plt.show()
```

```python
print('Number of elements of Xbox One sample:', xone.shape[0])
print('Number of elements of PC sample:', pc.shape[0])
```

- Распределения незначительно отклоняются от нормального, количество элементов привышает 30 - следовательно можно использовать парный т-тест;
- У нас нет информации о равенстве дисперсий генеральных совокупностей, из которых взяты выборки, поэтому поставил параметр equal_var=False;

```python
alpha = .05
results = st.ttest_ind(xone, pc, equal_var=False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('Reject hull hypothesis')
else:
    print('Cannot reject null hypothesis')
```

- Отвергнуть нулевую гипотезу не удалось, причем вероятность случайно получить такую разницу в пользовательских оценках довольно высокая;
- Посмотрим дополнительно боксплот и пойнтплот по нашим данным;

```python
xone = first_hypothesis.loc[first_hypothesis.platform == 'XOne', ['user_score', 'platform']]
pc = first_hypothesis.loc[first_hypothesis.platform == 'PC', ['user_score', 'platform']]

xone_pc_concat = pd.concat([xone, pc])

fig, ax = plt.subplots(figsize=(12, 6))
fig_1, ax_1 = plt.subplots(figsize=(6, 7))

sns.boxplot(x='user_score', 
            y='platform', 
            data=xone_pc_concat,
            ax=ax
           )
sns.pointplot(x='platform', 
              y='user_score', 
              data=xone_pc_concat,
              ax=ax_1,
              join=False,
              ci='sd'
             )

plt.show()
```

    - Нулевая гипотеза H0: Средние пользовательские рейтинги жанров action и sports **равны**;
    - Альтернативная гипотеза H1: Средние пользовательские рейтинги жанров action и sports **не равны**;
- Нам нужно проверить равенство средние двух выборок, т.е. нам нужно понять, принадлежат ли выборке одной генеральной совокупности, или двум разным;
- Подходящим методом будет парный т-тест, но для начала проверим распределение на нормальность и посмотрим количество элементов в выборках, чтобы понять, можно ли применять т-тест;

```python
# для проверки первой гипотезы уберем все пропуски и значения tbd
second_hypothesis = games_actual.copy(deep=True)
second_hypothesis = second_hypothesis[['genre', 'user_score']]

print('Dataset length BEFORE', first_hypothesis.shape[0])

second_hypothesis['user_score'].replace(-1, np.nan, inplace=True)
second_hypothesis.dropna(subset=['user_score'], inplace=True)

print('Dataset length AFTER', second_hypothesis.shape[0])
print('')
print('Number of NaNs: ')
display(second_hypothesis.isna().sum())
print('Number of tbd values:', second_hypothesis[second_hypothesis['user_score'] < 0]['user_score'].count())
```

```python
# создаем отдельные датасеты для двух платформ
action = second_hypothesis[second_hypothesis.genre == 'Action']['user_score']
sports = second_hypothesis[second_hypothesis.genre == 'Sports']['user_score']

# делаем стандартизацию
action_z = (action-np.mean(action))/np.std(action)
sports_z = (sports-np.mean(sports))/np.std(sports)

# выводим Q-Q графики
fig, ax = plt.subplots(figsize=(10, 6))
fig_1, ax_1 = plt.subplots(figsize=(10, 6))

st.probplot(action_z, dist='norm', plot=ax)
st.probplot(sports_z, dist='norm', plot=ax_1)

ax.set_title('Normal Q-Q plot Action')
ax_1.set_title('Normal Q-Q plot Sports')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.distplot(action, 
             label='Xbox One',
             bins=15
            )
sns.distplot(sports, 
             label='PC',
             bins=15
            )

plt.ylim(0, 0.4)
plt.axvline(action.mean(), 
            c='tab:blue', 
            linewidth=1.5, 
            linestyle='--'
           )
plt.axvline(sports.mean(), 
            c='peru', 
            linewidth=1.5, 
            linestyle='--'
           )
ax.text(action.mean()+.1, 
        0.32, 
        'Action mean', 
        rotation=90
        )
ax.text(sports.mean()-.2, 
        0.33, 
        'Sports mean', 
        rotation=90
        )


plt.legend(loc=2)
plt.show()
```

```python
print('Number of elements of Xbox One sample:', action.shape[0])
print('Number of elements of PC sample:', sports.shape[0])
```

- Распределения незначительно отклоняются от нормального, количество элементов привышает 30 - следовательно можно использовать парный т-тест;

```python
alpha = .05
results = st.ttest_ind(action, sports, equal_var=False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('Reject hull hypothesis')
else:
    print('Cannot reject null hypothesis')
```
