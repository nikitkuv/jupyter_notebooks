# Выбор локации для скважины

Предоставлены пробы нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов.

**Цель:** Построить модель машинного обучения, которая поможет определить регион, где добыча принесёт наибольшую прибыль. Проанализировать возможную прибыль и риски техникой *Bootstrap.* 

**Целевой признак модели:** объём запасов в скважине, тыс. баррелей (product).



Шаги для выбора локации:
- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.

Условия задачи:
- Для обучения модели подходит только линейная регрессия (остальные — недостаточно предсказуемые).
- При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
- Бюджет на разработку скважин в регионе — 10 млрд рублей.
- При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.
- После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.

# Описание данных

- id — уникальный идентификатор скважины;
- f0, f1, f2 — параметры нефти;
- product — объём запасов в скважине (тыс. баррелей).

## Обучение и проверка модели

### Описание подхода к созданию модели

- Целевой признак (product) является количественной переменной. Для ее предсказания будет использоваться алгоритмы линейной регрессии LinearRegression
- Т.к. признаки не содержат явных выбросов, для оценки качества моделей можно использовать MSE (RMSE), а также дополнительно будем считать R2
- Делить данные будем на обучающую и тестовую выборки в отношении 75:25
- Алгоритм у нас линейный и distance-based, к тому же все признаки количественные - следовательно после разбиения данных на обучающую и тестовую, но перед кросс-валидацией, будем делать стандартизацию признаков с использованием объекта StandardScaler
- Обучающую выборку будем использовать в кросс-валидации

```python
seed = 24
cv_k = 5

x_1 = geo_1.drop(['id', 'product'], axis=1) # датафрейм с признаками
y_1 = geo_1.loc[:, 'product'] # серия с целевым признаком

x_2 = geo_2.drop(['id', 'product'], axis=1) # датафрейм с признаками
y_2 = geo_2.loc[:, 'product'] # серия с целевым признаком

x_3 = geo_3.drop(['id', 'product'], axis=1) # датафрейм с признаками
y_3 = geo_3.loc[:, 'product'] # серия с целевым признаком

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(x_1, 
                                                            y_1, 
                                                            test_size=0.25, 
                                                            random_state=seed, 
                                                           )
x_train_2, x_test_2, y_train_2, y_test_2 = train_test_split(x_2, 
                                                            y_2, 
                                                            test_size=0.25, 
                                                            random_state=seed, 
                                                           )
x_train_3, x_test_3, y_train_3, y_test_3 = train_test_split(x_3, 
                                                            y_3, 
                                                            test_size=0.25, 
                                                            random_state=seed, 
                                                           )
```

### LinearRegression

#### Регион 1

```python
scaler_linreg = StandardScaler()
linreg = LinearRegression()

pipe_linreg_1 = Pipeline([('scaler', scaler_linreg), ('linreg', linreg)])

train_val_score_linreg_1 = cross_validate(pipe_linreg_1,
                                          x_train_1, 
                                          y_train_1, 
                                          cv=cv_k, 
                                          scoring=('r2', 'neg_mean_squared_error'),
                                          return_train_score=True
                                          )

pipe_linreg_1.fit(x_train_1, y_train_1)
y_pred_1 = pipe_linreg_1.predict(x_test_1)

r2_test_score_linreg_1 = round(r2_score(y_test_1, y_pred_1), 3)
rmse_test_score_linreg_1 = round(mean_squared_error(y_test_1, y_pred_1)**.5, 3)

r2_val_score_linreg_1 = round(train_val_score_linreg_1['test_r2'].mean(), 3)
rmse_val_score_linreg_1 = round(((-train_val_score_linreg_1['test_neg_mean_squared_error'])**.5).mean(), 3)

r2_train_score_linreg_1 = round(train_val_score_linreg_1['train_r2'].mean(), 3)
rmse_train_score_linreg_1 = round(((-train_val_score_linreg_1['train_neg_mean_squared_error'])**.5).mean(), 3)

product_std_1 = np.std(geo_1['product']).round(3)

print(f'R2 test: {r2_test_score_linreg_1}')
print(f'RMSE test: {rmse_test_score_linreg_1}')
print('')
print(f'R2 valid: {r2_val_score_linreg_1}')
print(f'RMSE valid: {rmse_val_score_linreg_1}')
print('')
print(f'R2 train: {r2_train_score_linreg_1}')
print(f'RMSE train: {rmse_train_score_linreg_1}')
print('')
print(f'Product std: {product_std_1}')
```

- Ошибка по тестовой выборке = ошибке по валидационной и почти = ошибке по тестовой => модель не страдает от переобучения или недообучения
- По модели мы в среднем ошибаемся на 38 тыс. баррелей (RMSE), при это среднее отклонение данных от среднего, или величина стандартного отклонения, равна 44 тыс. баррелей. Таким образом, ошибка немного меньше стандартного отклонения, поэтому по метрике RMSE работу модели можно считать удовлетворительной.
- Но при этом довольно низкие значения коэффициента детерминации по всем выборкам

#### Регион 2

```python
scaler_linreg = StandardScaler()
linreg = LinearRegression()

pipe_linreg_2 = Pipeline([('scaler', scaler_linreg), ('linreg', linreg)])

train_val_score_linreg_2 = cross_validate(pipe_linreg_2,
                                          x_train_2, 
                                          y_train_2, 
                                          cv=cv_k, 
                                          scoring=('r2', 'neg_mean_squared_error'),
                                          return_train_score=True
                                          )

pipe_linreg_2.fit(x_train_2, y_train_2)
y_pred_2 = pipe_linreg_2.predict(x_test_2)

r2_test_score_linreg_2 = round(r2_score(y_test_2, y_pred_2), 3)
rmse_test_score_linreg_2 = round(mean_squared_error(y_test_2, y_pred_2)**.5, 3)

r2_val_score_linreg_2 = round(train_val_score_linreg_2['test_r2'].mean(), 3)
rmse_val_score_linreg_2 = round(((-train_val_score_linreg_2['test_neg_mean_squared_error'])**.5).mean(), 3)

r2_train_score_linreg_2 = round(train_val_score_linreg_2['train_r2'].mean(), 3)
rmse_train_score_linreg_2 = round(((-train_val_score_linreg_2['train_neg_mean_squared_error'])**.5).mean(), 3)

product_std_2 = np.std(geo_2['product']).round(3)

print(f'R2 test: {r2_test_score_linreg_2}')
print(f'RMSE test: {rmse_test_score_linreg_2}')
print('')
print(f'R2 valid: {r2_val_score_linreg_2}')
print(f'RMSE valid: {rmse_val_score_linreg_2}')
print('')
print(f'R2 train: {r2_train_score_linreg_2}')
print(f'RMSE train: {rmse_train_score_linreg_2}')
print('')
print(f'Product std: {product_std_2}')
```

## Расчёт прибыли и рисков 

### Расчет прибыли по данным

- Напишем функцию для расчета прибыли

```python
def profit_calculation(target, predictions):
    predictions_sorted = predictions.sort_values(ascending=False)
    selected_points = target[predictions_sorted.index][:SELECTION_SPACE]
    product = selected_points.sum()
    revenue = product * HC_PRICE
    cost = well_cost * SELECTION_SPACE
    return round(revenue - cost)
```

```python
profit_results = []

for i in [(y_test_1, y_pred_1), (y_test_2, y_pred_2), (y_test_3, y_pred_3)]:
    profit_results.append(profit_calculation(i[0], i[1]))
    
for index, i in enumerate(profit_results):
    print(f'Region {index+1} profit: {i} thousand rubles')
```

- По расчетам моделей район 1 является районом с наибольшими запасами и соответственно прибылью

### Расчет bootstrap распределения прибыли и рисков

- С помощью бутстрепа получим распределения прибыли для каждого региона

```python
profit_bs_1 = np.empty(1000)
profit_bs_2 = np.empty(1000)
profit_bs_3 = np.empty(1000)

for i in [(y_test_1, y_pred_1, profit_bs_1), 
          (y_test_2, y_pred_2, profit_bs_2), 
          (y_test_3, y_pred_3, profit_bs_3)
         ]:
    for k in range(1000):
        target_subsample = i[0].sample(ANALYSIS_SPACE, replace=True, random_state=state)
        pred_subsample = i[1][target_subsample.index]
        prof = profit_calculation(target_subsample, pred_subsample)
        i[2][k] = prof
```

```python
fig, axs = plt.subplots(figsize=(12,8))

sns.distplot(profit_bs_1, ax=axs)
sns.distplot(profit_bs_2, ax=axs)
sns.distplot(profit_bs_3, ax=axs)

plt.axvline(profit_bs_1.mean(), linewidth=1.5, linestyle='--')
plt.axvline(profit_bs_2.mean(), c='tab:orange', linewidth=1.5, linestyle='--')
plt.axvline(profit_bs_3.mean(), c='forestgreen', linewidth=1.5, linestyle='--')

plt.text(profit_bs_1.mean()+10000, 0.0000015, f'Mean region 1: {int(profit_bs_1.mean().round())}', c='tab:blue')
plt.text(profit_bs_2.mean()+10000, 0.00000195, f'Mean region 2: {int(profit_bs_2.mean().round())}', c='tab:orange')
plt.text(-90000, 0.00000162, f'Mean region 3: {int(profit_bs_3.mean().round())}', c='forestgreen')

plt.legend(['Region 1', 'Region 2', 'Region 3'])
plt.title('Bootstrap profit distributions')
plt.xlabel('Profit, thousands rubles')

plt.show()
```

- Bootstrap распределения прибыли показывают, что **самое большое среднее значение прибыли ожидается в регионе 2**, наименьшее - в регионе 3
- При этом распределение региона 2 имеет наименьшее стандартное отклонение, а значит и в среднем наименьший разброс значений вокруг среднего, а стало быть, в среднем более точную оценку прибыли

- Посчитаем 95% доверительные интервалы для каждого из распределений прибыли и посмотрим на боксплоты этих распределений, расположение границы "усов" в пределах 95% доверительного интервала

```python
profit_bs_1 = pd.Series(profit_bs_1)
profit_bs_2 = pd.Series(profit_bs_2)
profit_bs_3 = pd.Series(profit_bs_3)
```

```python
fig, axs = plt.subplots(figsize=(8,10))

ci_values = []

for i in [profit_bs_1, profit_bs_2, profit_bs_3]:
    ci_values.append(st.t.interval(.95, len(i)-1, i.mean(), i.sem()))
    
for index, i in enumerate([profit_bs_1, profit_bs_2, profit_bs_3]):
    print(f'Доверительный интервал региона {index+1}: {i.quantile(.025).round()} - {i.quantile(.975).round()}')

sns.boxplot(x='regions',
            y='profit',
            data=pd.DataFrame({'profit':pd.concat([profit_bs_1, profit_bs_2, profit_bs_3],axis=0),
                               'regions':np.repeat(['1', '2', '3'],[len(profit_bs_1),len(profit_bs_2), len(profit_bs_3)]
                                                )
                              }
                             ),
            whis=[2.5, 97.5]
           )

plt.title('Bootstrap profit by regions, whiskers are in 95% CI')
plt.show()
```

- Доверительный интервал уже всего у региона 2, что говорит о более точной оценке
- Медианное значение прибыли также выше в регионе 2
- **Только для региона 2 значения 95% доверительного интервала полностью положительные**, для остальных регионов часть интервала залезает в отрицательную область, что говорит о вероятности получить убыток для этих регионов

- Расчитаем риски получить убыток для каждого из регионов
- Для этого аппроксимируем распределения прибыли по регионам нормальным распределением: используя среднее и стандартное отклонение выборок, посчитаем функцию плотности вероятности для каждой выборки, а затем на ее основе посчитаем вероятность получить прибыль 0 или меньше - это и будет риск получить убыток

```python
risk_probs = []

for i in [profit_bs_1, profit_bs_2, profit_bs_3]:
    risk_probs.append(st.norm(i.mean(), i.std()).cdf(0))
    
for index, i in enumerate(risk_probs):
    print('Риск убытков для региона {}: {:.2%}'.format(index+1, i))
```

### Вывод

- Наибольшее среднее значение прибыли ожидается в регионе 2
- Наиболее точная оценка распределения прибыли наблюдается в регионе 2
- Наименьший риск убытков, и единственное значение, ниже требуемых 2.5% - в регионе 2
