## Методология решения задачи

В задачах прогноза оттока клиентов выделяются две основные группы с точки зрения определения метки целевого признака для каждого клиента: **контрактовый** и **внеконтрактовый**.

Конктрактовый отток основан на том, что клиент явно информирует о прекращении пользованием продукта (банки, телеком компании и т.д.). Внеконтрактовый отток характеризуется тем, что сам факт оттока не зафиксирован явно, как, например в случае наших данных: мы не знаем точно, связана ли неактивность клиента с тем, что он перестал пользоваться нашими услугами.

Существуют 3 основных подхода для оценки того, ушел клиент или нет, во внеконтрактовых условиях:
1. Создаем целевой признак на основе знаний из области - здесь нам должен помочь бизнес: клиента можно считать ушедшим, если с момента последней покупки прошло 30 дней. У нас, к сожалению, в данный момент нет возможности получить подобные ориентиры от бизнеса.
2. Вероятностный подход: вероятность того, что клиент все еще пользуется продуктом, высока сразу после приобритения товара и постепенно падает со временем. Клиенты делают покупки с разной частотой, иными словами, имеют разное время между покупками, но в целом новая покупка происходит в районе среднего времени между покупок, таким образом, если время между покупками превышает среднее - вероятность того, что клиент ушел, возрастает. Используемые модели и подход отличаются от классических алгоритмов машинного обучения. В данном подходе нам понадобятся только данные по транзакциям клиентов. Для предсказания вероятности ухода клиента нам также нужны лишь данные по транзакциям.
3. Создаем целевой признак на основе предсказаний одной модели (например, вероятностной), а затем, используя данные все имеющиеся данные (не только транзакции, как в вероятностном подходе), строим вторую модель с использованием классических табличных алгоритмов машинного обучения. В данном случае мы сгруппируем и объединим данные по клиентам по всем таблицам. Здесь для предсказания вероятности ухода клиента нужны данные как по транзакциям, так и по другим таблицам, в зависимости от выбранных признаков.

## Подход №1: Beta Geometric/Negative Binomial Distribution Model (BG/NBD)

Для начала попробуем использовать **вероятностный подход** и основанную на нем модель **Beta Geometric/Negative Binomial Distribution Model (BG/NBD)** из библитеки `lifetimes`.

Предсказания данной модели основаны на метриках RFM (recency, frequency, monetary value, а также age или longevity). Концепция RFM подразумевает, что клиенты, сделавшие покупки недавно, покупающие чаще, и тратящие больше, являются наиболее вероятно действующими клиентами: вероятность их оттока наименьшая.

**Recency** - временной интервал между первой и последней покупкой.

**Frequency** - количество повторных покупок после первой транзакции.

**Monetary Value** - усредненнная выручка на покупку с клиента.

**Age T** - временной интервал между первой покупкой и последней датой горизонта исследований (в нашем случае - последняя дата в таблице транзакций)

### Подготовка данных

Входными данными для модели BG/NBD являются метрики RFM: Recency, Frequency, Monetary Value и Age T, которые являются агрегирующими по всем уникальным клиентам.

Для расчета Recency, Frequency и Age T нам нужны только данные по датам транзакций - *chq_date*.

Monetary Value у нас уже есть - это столбец *sales_sum*.

Столбец с айди клиентами - *client_id*.

Таким образом, здесь нам нужен лишь датафрейм с транзакциями.

Помимо всего прочего, мы создадим новый датафрейм только со столбцами *chq_date*, *client_id* и *sales_sum*.

Все остальные столбцы, не участвующие в расчетах RFM признаков, мы уберем далее конкретно для этого подхода.

P.S. Подготовка всех остальных столбцов описана в главе 6.1.

### Учет дисбаланса классов целевого признака

Для учета дисбаланса классов целевого признака воспользуемся техникой оверсемплинга и функцией SMOTE.

Объединим стандартизацию количественных признаков, оверсемплинг, обучение и подбор гиперпараметров с помощью GridSearch и оценку качества модели по ROC AUC на тестовой выборке в одну функцию, которая будет возвращать лучшие гиперпараметры, значения метрик, предсказанные значения и сами модели.

```python
def imb_model_score_pipeline(X_train, 
                             y_train, 
                             X_test, 
                             y_test, 
                             model, 
                             seed, 
                             gs_params, 
                             cv_k, 
                             oversampling=True, 
                             scale=True, 
                             randomized_cv=False
                            ):
    """
    
    """
    
    if scale:
        if oversampling==False:
            imb_pipeline = make_pipeline(StandardScaler(), 
                                         model
                                        )
        else:
            imb_pipeline = make_pipeline(StandardScaler(),
                                         SMOTE(random_state=seed), 
                                         model
                                        )
    else:
        if oversampling==False:
            imb_pipeline = model
        else:
            imb_pipeline = make_pipeline(SMOTE(random_state=seed), 
                                         model
                                        )
    
    if randomized_cv:
        grid_imb = RandomizedSearchCV(imb_pipeline, 
                                      param_grid=gs_params, 
                                      cv=cv_k, 
                                      scoring='roc_auc', 
                                      return_train_score=True,
                                      #refit=scoring
                                     )
    else:
        grid_imb = GridSearchCV(imb_pipeline, 
                                param_grid=gs_params, 
                                cv=cv_k, 
                                scoring='roc_auc', 
                                return_train_score=True,
                                #refit=scoring
                               )
    
    grid_imb.fit(X_train, y_train)
    
    best_model = grid_imb.best_estimator_
    best_params = grid_imb.best_params_
    
    y_pred_probs = best_model.predict_proba(X_test)
    y_pred_probs_ones = best_model.predict_proba(X_test)[:, 1]
    
    y_pred = best_model.predict(X_test)
    
    test_roc_auc = roc_auc_score(y_test, y_pred_probs_ones).round(4)
    val_roc_auc = np.nanmean(grid_imb.cv_results_['mean_test_score']).round(4)
    train_roc_auc = np.nanmean(grid_imb.cv_results_['mean_train_score']).round(4)
    
    return best_params, test_roc_auc, val_roc_auc, train_roc_auc, y_pred, y_pred_probs, best_model
```
